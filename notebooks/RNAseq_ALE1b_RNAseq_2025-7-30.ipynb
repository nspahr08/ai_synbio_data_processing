### Trimming

Based on FastQC reports, Nextera Transposase Adapters are present at the end of many reads.

```python
import subprocess
from binfo_utils import create_manifest

trimmed = os.path.join(base_dir, 'trimmed')
os.makedirs(trimmed, exist_ok=True)

manifest = create_manifest(received)
for index, row in manifest.iterrows():

    read1 = str(os.path.basename(row['fwd_fastq'])).replace('.fastq.gz', '_trimmed.fastq.gz') 
    read2 = str(os.path.basename(row['rvs_fastq'])).replace('.fastq.gz', '_trimmed.fastq.gz')
    
    cutadapt_cmd = [
        'cutadapt',
        '-j', '10',
        '-q', '15, 15',  # quality cut-off, trimmed from both reads
        # '-b', 'TCGTCGGCAGCGTCAGATGTGTATAAGAGACAG',  # Read 1 Nextera Transposase adapter, not it
        # '-B', 'GTCTCGTGGGCTCGGAGATGTGTATAAGAGACAG',  # Read 2 Nextera Transposase adapter
        '-b', 'ACTGTCTCTTATACACATCT',  # Illumina Stranded mRNA and Illumina Stranded Total RNA Adapter Trimming
        '-B', 'ACTGTCTCTTATACACATCT', 
        '--revcomp',
        # '-a', 'AGATCGGAAGAGCACACGTCTGAACTCCAGTCA',  # TrueSeq Read1 adapter, not it
        # '-A', 'AGATCGGAAGAGCGTCGTGTAGGGAAAGAGTGT',  # TrueSeq Read2 adapter
        '-b', 'CTGTCTCTTATACACATCT',  # Nextera mate-pair adapter
        '-B', 'AGATGTGTATAAGAGACAG',  # Nextera mate-pair adapter
        # '--poly-a',
        '--minimum-length',  '20',
        '-o', os.path.join(trimmed, read1),
        '-p', os.path.join(trimmed, read2),
        row['fwd_fastq'],
        row['rvs_fastq'],
    ]

    subprocess.run(cutadapt_cmd, check=True, cwd=trimmed)
```

### Map RNAseq reads to their reference & count features with FADU

References were created as follows:
1. Chanty made a reference for ACN2821 (2 DAHP synthases missing, dgoA* inserted, chromosomal mutations added).
2. I made versions where dgoA* is replaced with dgoA-Ec or dgoA-Best.
3. Changed the feature type for dgoA from 'CDS' to 'gene'. (So that it can be found by FADU). 
4. Exported these SnapGene files as .gbk files.

```python
import os
from mapping import map_and_feature_count

ACN2853_samples = manifest.loc[manifest['sample_name'].str.contains('ACN2853')]
ACN3210_samples = manifest.loc[manifest['sample_name'].str.contains('ACN3210')]

mapped = os.path.join(base_dir, 'mapped')
os.makedirs(mapped, exist_ok=True)

for index, row in ACN2853_samples.iterrows():
    print(f"Processing sample {row['sample_name']}")
    fwd_reads = row['fwd_fastq']
    rvs_reads = row['rvs_fastq']
    sample_name = row['sample_name']
    reference = '/Users/nataschaspahr/Reference_DBs/genomes/ACN2821_CDM_dgoAEcoli_NSS_dgoAasGene.gbk'
    output_dir = os.path.join(mapped, row['sample_name'])
    os.makedirs(output_dir, exist_ok=True)
    
    map_and_feature_count(fwd_reads, rvs_reads, reference, output_dir, keep_index=False, sample_name=sample_name)

for index, row in ACN3210_samples.iterrows():
    print(f"Processing sample {row['sample_name']}")
    fwd_reads = row['fwd_fastq']
    rvs_reads = row['rvs_fastq']
    sample_name = row['sample_name']
    reference = '/Users/nataschaspahr/Reference_DBs/genomes/ACN2821_CDM_dgoABest_NSS_dgoAasGene.gbk'
    output_dir = os.path.join(mapped, row['sample_name'])
    os.makedirs(output_dir, exist_ok=True)
    
    map_and_feature_count(fwd_reads, rvs_reads, reference, output_dir, keep_index=False, sample_name=sample_name)
```

### Create feature count summary for all samples

```python
import pandas as pd
import glob

df = None

for folder in sorted(os.listdir(mapped)):
    count_files = glob.glob(os.path.join(mapped, folder, '*.txt'))
    base_filename = os.path.basename(count_files[0])
    sample_name = base_filename.split('.sorted.counts.txt')[0]
    counts = pd.read_csv(count_files[0], sep='\t')
    counts.drop(columns=['num_alignments', 'counts'], inplace=True)
    if df is None:
        df = counts
        df.rename(columns={'tpm': sample_name}, inplace=True)
    else:
        counts.rename(columns={'tpm': sample_name}, inplace=True)
        counts.drop(columns=['uniq_len'], inplace=True)
        df = pd.merge(df, counts, on='featureID', how='outer')

df.to_csv(os.path.join(mapped, 'TPM_counts.csv'), index=False)
```
